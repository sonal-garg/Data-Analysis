### Data Cleaning (Replace this with more specific notes!)

#### Removing any duplicate rows

```python
# After discussing the structure of the data and any problems that need to be
#   cleaned, perform those cleaning steps in the second part of this section.

# Checking for duplicate rows
df[df.duplicated(keep=False)]
```

```python
# Removing the duplicate row
df.drop_duplicates(inplace=True)
```

```python
#Checking
df.shape
```

#### Removing unnecessary columns

```python
df.head(2)
```

Looking at these columns, I feel there is no use of maintaining two id columns. Let's just keep one of them. So I will get rid of imdb_id column.

I feel having tagline and overview column seems a little redundant. Anyways both of these columns do not provide any significant data that can be analysed.So, I decide to remove both of these columns.

Since we already have the budget and revenue stated, we can remove budget_adj and revenue_adj column.

There is a popularity score column. So, we dont neccesarily need the vote_vount and vote_average column. I dont think I will be using these 2 columns for any analysis.

So, overall lets remove 7 columns namely 'imdb_id,tagline,overview,budget_adj,revenue_adj,vote_count,vote_average'

```python
col=['imdb_id','tagline','overview','budget_adj','revenue_adj','vote_count','vote_average']
df.drop(labels=col,axis=1,inplace=True)
```

```python
#Checking
df.shape
```

#### Checking for columns with null values

On checking for null values in various colums, I came across certain rows, where budget and revenue were zero. Also, there were certain columns where the budeget was zero, but revenue was generated. However, as we have seen before, there are no null values in the release datecolumn. That means all these movies do have a release date.Lets look into this kind of data further.

```python
df.query('budget == 0').count()['id']
```

```python
df.query('revenue==0').count()['id']
```

This is too much data to be removed.So, I decide to keep these rows as it is. I will just replace these values of zero with null.

```python
df['budget'].replace(0,None,inplace=True)
df['revenue'].replace(0,None,inplace=True)
```

```python
#Checking
print(df['budget'].isnull().sum())
print(df['revenue'].isnull().sum())
```

```python
df.isnull().sum()
```

There are some small number of null values present in cast,director and genres. Lets check these out. Again, since homepage and keywords have larger number of null values. I am going to keep them as it is.

```python
df[df.genres.isnull()]
```

Since, I am going to analyse the data genre wise, I will remove the rows where genre is null. anyways, the count is very small(23 rows).

I also decided to remove the null rows for column 'director' and 'cast'.The count again was very small and I might decide to look at the data director wise.

```python
col=['genres','cast','director']
df.dropna(subset=col,inplace=True)
```

```python
#Checking
df.isnull().sum()
```

```python
df.info()
```

#### Correction of datatypes for 'release_date'

```python
#Converting release dat from string to datetime
df['release_date']=pd.to_datetime(df['release_date'])
```

```python
#Checking
df.info()
```

## Exploratory Data Analysis


### Research Question 1 (What is the most popular genre of movies?)

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.

#Splitting the genres seprated by |.
genres_split=df['genres'].str.cat(sep='|')
genres_split = pd.Series(genres_split.split('|'))

#Couting by genres
genre_count = genres_split.value_counts(ascending = False)
genre_count
```

```python
genre_count.plot(kind= 'barh',figsize=(12,5))

plt.title("Genre wise distribution of movies",fontsize=14)
plt.xlabel('Number Of Movies',fontsize=12)
plt.ylabel("Genres",fontsize= 12);
```

Clearly, over the years **drama has been the most popular category of movies** followed by comedy.

### Research Question 2  (How are budget and revenue related?)

```python
# Continue to explore the data to address your additional research
#   questions. Add more headers as needed if you have more questions to
#   investigate.

#Sorting the movies by budget
df[df['budget'].notnull()][['original_title','budget','release_year']].sort_values('budget',ascending=False).head(5)
```

**Hubble 3D** and **The Warrior's Way** are the movies made on maximum budget. They both released in 2010

```python
df[df['revenue'].notnull()][['original_title','revenue','release_year']].sort_values('revenue',ascending=False).head(5)
```

So, **Avatar** has been the most revenue generating movie. It released in 2009.**Star Wars: The Force Awakens** came close in 2015 but did not surpass the record.

```python
print("The average budget of movies over the years is:",df['budget'].mean())
print("The average revenue earned by movies over the years is:",df['revenue'].mean())
```

```python
#plotting budget and revenue over the years
budget_yearly=df[df['budget'].notnull()][['release_year','budget']].groupby('release_year').mean()
revenue_yearly=df[df['revenue'].notnull()][['release_year','revenue']].groupby('release_year').mean() 

plt.figure(figsize=(12,5))
plt.plot(budget_yearly,label='Mean budget trends')
plt.plot(revenue_yearly,label='Mean revenue trends')

plt.xlabel("Year of release",fontsize=12)
plt.ylabel("Mean value",fontsize=12)
plt.title("Mean Budget/Revenue trends over the years",fontsize=14)
plt.legend(loc='upper right')
plt.show();
```

So, as per the above plot,we find movies on an average have always returned better revenue than there budget. But does, that mean there have been no losses incurred? That can't be possible.Lets try to find out about not so profitable movies.

### Research Question 3  (Which is the most profitable and least profitable movie of all time?)

To calculate profit of a movie, we will create a new column, which calculates profit based on revenue and budget.

```python
#Insering a 'profit' column after 'budget' and 'revenue' col
df.insert(4,'profit',df['revenue']-df['budget'])
df.head(2)
```

```python
#checking for flop movies
df.query('profit<0')
```

So, there are 4517 movies that have incurred losses.

```python
#Checking for highest profit making movie
df[df['profit'] == df['profit'].max()]
```

```python
# checking for lowest profit making movie
df[df['profit'] == df['profit'].min()]
```

So, there are no changed patterns here, most profitable movie remains **Avatar** and least profitable movies remain **The Warrior's Way** and **Hubble 3D**. There is no new information so far here.Lets check is there any trend for movies that have incurred losses.
**Which year has had the most number of losses?**

### Research Question 4  (Which year has seen maximum losses and what are the average number of flop movies per year?)

```python
#Analysing flop movies
movies_losses=df['profit'][df['profit']<0]
movies_release=df['release_year'][df['profit']<0]

plt.figure(figsize=(12,5))

plt.bar(movies_release,movies_losses)
plt.gca().invert_yaxis()
plt.xlabel('Year of Release',fontsize=12)
plt.ylabel('Losses incurred',fontsize=12)
plt.title('Movies that incurred losses over the years',fontsize=14)

plt.show();
```

**2010 has been the least profitable year so far followed by 2006**.There is a huge gap between data for 2010 and other years. this is understandable as the most loss incurring movies released in this year itself(The Warrior's Way and Hubble 3d).
Lets check similar data specifically for 2006

```python
df[df['release_year']==2006].sort_values(by='profit')
```

So there are 3 movies that have been huge flops this year and that is why, this is the second poor year.

```python
#Computing loss % of movies each year.

df_losses=df[df["profit"] <0].groupby("release_year")
df_losses['profit'].count()

df_total=df.groupby("release_year")
df_total['profit'].count()

print("On an average below are the % of movies that flop each year")
print(round((df_losses['profit'].count()/df_total['profit'].count())*100,2))
```

```python
prop_losses=(df_losses['profit'].count()/df_total['profit'].count())*100

plt.figure(figsize=(12,5))

plt.plot(prop_losses)
plt.xlabel('Year of Release',fontsize=12)
plt.ylabel('Loss %',fontsize=12)
plt.title('Loss percentage over the years',fontsize=14)

plt.show()
```

So the rate of movies being flopped has been gradually increasing after 1980. Though there was an unusual peak in 1970 and 1996.

### Research Question 5  ( Which has been the most succesful year so far?)

```python
movie_yearlyprofit=df.groupby('release_year')['profit'].sum().sort_values(ascending=False)
movie_yearlyprofit.head(8)
```

So **2013 has been the most succesful year in terms of profit**. Surprisingly, its not 2009 when Avatar was released.

```python
plt.figure(figsize=(12,5))
plt.plot(movie_yearlyprofit)
plt.xlabel('Year of release',fontsize=12)
plt.ylabel('Profit',fontsize=12)
plt.title('Profitability over the years',fontsize=14)
plt.show()
```

Profitability in movies have been increasing year by year.

### Research Question 6  (Which actor/actress has done most number of succesful movies?)

Moving forward in the anlaysis, to measure the success of movies, we will be creating another dataframe, which has the dataset of only successful movies i.e movies that have made any profit.For that,I am putting movies in this dataset with profit>0

```python
# Only considering profitable movies.

df_profitable_movies=df[df['profit']>0]
```

```python
df_profitable_movies['budget'].mean()
```

```python
df_profitable_movies['revenue'].mean()
```

```python
# Splitting the cast column
cast_data = df_profitable_movies['cast'].str.cat(sep = '|')    
cast_data = pd.Series(cast_data.split('|'))

#Counting by cast
cast_count = cast_data.value_counts(ascending = False)
cast_count.head(5)
```

**Robert De Niro has done the most number of succesful movies**

### Research Question 7  (How are runtimes of movies changing over the years? Is there any corelation between runtime and popularity?)

```python
yearly_runtimes=df_profitable_movies.groupby('release_year')['runtime'].mean().sort_index()
yearly_runtimes
```

```python
plt.figure(figsize=(12,5))
plt.plot(yearly_runtimes)
plt.title('Average Runtime of Movies over the Years',fontsize=14)
plt.xlabel('Year of release',fontsize=12)
plt.ylabel('Runtime',fontsize=12)
plt.show()
```

Clearly the average runtime of movies have decreased significantly from 122mins in 1962 to 99 minutes in 2015.As the profotability has also increased over time we can say, that now shorter duration movies are more liked by people.

### Research Question 8  (What are the popularity trends over the years?)

```python
#Checking for the film with max popularity score
df_profitable_movies[df_profitable_movies['popularity']==df_profitable_movies['popularity'].max()]
```

**Jurassic World is the most popular movie of all time**

```python
#Checking for the film with min popularity score
df_profitable_movies[df_profitable_movies['popularity']==df_profitable_movies['popularity'].min()]
```

The hospital is the least popular movie, though it has earned a decent amount of profit.

```python
df_profitable_movies.groupby('release_year')['popularity'].mean().plot()
```

Populartity of successful movies has been increasing with years.

### Research Question 9  (Which month had the most number of releases ?)

```python
#Extracting month
df['month_str'] = df['release_date'].dt.strftime('%b')
```

```python
#Calulating number of releases in each month.
monthly_release=df['month_str'].value_counts()
monthly_release.index = pd.CategoricalIndex(monthly_release.index, categories=['Jan', 'Feb', 'Mar', 'Apr','May','Jun', 'Jul', 'Aug','Sep', 'Oct', 'Nov', 'Dec'])

#Sorrting the data as sequence of months
monthly_release = monthly_release.sort_index()
monthly_release
```

**Septemeber and October seems to be preferred months for movie release**

```python
plt.figure(figsize=(12,8))
monthly_release.plot.bar(alpha=0.4,color='orange')

plt.title('Number of movie releases by month',fontsize=14)
plt.xlabel('Month of release',fontsize=12)
plt.ylabel('Number of movies',fontsize=12)
plt.show()
```

### Research Question 10  (Which month has been the most profitable month over the years ?)

```python
#Grouping by monthly profit
monthly_profit=df[df["profit"]>0].groupby("month_str")['profit'].mean()

monthly_profit.index = pd.CategoricalIndex(monthly_profit.index, categories=['Jan', 'Feb', 'Mar', 'Apr','May','Jun', 'Jul', 'Aug','Sep', 'Oct', 'Nov', 'Dec'])

#Sorting as a sequence of month
monthly_profit = monthly_profit.sort_index()
monthly_profit
```

```python
plt.figure(figsize=(12,8))
monthly_profit.plot.bar(alpha=0.4,color='blue')

plt.title('Mean monthly revenue over the years',fontsize=14)
plt.xlabel('Month of release',fontsize=12)
plt.ylabel('Mean revenue',fontsize=12)
plt.show()
```

**June** has been the most successful month over the years. Films released in May/June have earned highest profits.

### Research Question 11  (Which director has made the most number of profitable movies ?)

```python
director_profit=df_profitable_movies.groupby('director')['id'].count().sort_values(ascending=False)
director_profit.head(10)
```

```python
plt.figure(figsize=(12,8))
director_profit.head(10).plot.bar()
plt.title('Mean monthly revenue over the years',fontsize=14)
plt.xlabel('Month of release',fontsize=12)
plt.ylabel('Mean revenue',fontsize=12)
plt.show()
```

**Clint Eastwood,Woody Allen are the directors that have made most number of profitable movies** closely followed by Steven Spielberg

### Research Question 12  (Which production company has been the most successful?)

```python
#Splitting the column'production_company'
prod_split=df_profitable_movies['production_companies'].str.cat(sep='|')
prod_split = pd.Series(prod_split.split('|'))

prod_count = prod_split.value_counts(ascending = False)
prod_count.head(10)
```

```python
plt.figure(figsize=(12,5))
prod_count.head(10).plot(kind='barh')

plt.title('Production wise distribution of movies ')
plt.xlabel('Number Of Movies')
plt.ylabel('Production companies')
plt.show()
```

**Universal Pictures** releases the most number of profitable movies followed by warner Bros.

## Conclusions

After analysing various parameters of the movie data, I have come up with the following conclusion:

Following parameters ensure for a movie to be succesful:

-  Genre: Genre of the movie should be either Drama or Comedy, preferrably Drama.
- Successful cast : Following are the top 3 most succesful cast members: Robert De Niro,Samuel L. Jackson,Tom Hanks.
- Smaller runtime: Runtime of the movie should be within 100 minutes.
- Good production house : Universal Pictures have produced maximum number of hits.
- Great Director: Top 3 most succeful directors are: Clint Eastwood,Woody Allen,Steven Spielberg.
- Appropriate time of release : Months June and May have recorded the maximum profit over the years.

Following are the trends observed over the years:

- Avatar is most profitable movie.
- The Warrior's Way and Hubble 3D are the least profitable movies.
- 2010 has been the least profitable year so far followed by 2006.
- 2013 has been the most successful year so far.
- September followed by october has seen the most number of releases over the years.
- Hubble 3D and The Warrior's Way are the movies made on maximum budget. They both released in 2010
- So, Avatar has been the most revenue generating movie. It released in 2009.Star Wars: The Force Awakens came close in 2015 but did not surpass the record
- Jurassic park is the most popular movie of all time.
- The Hospital is the least popular movie.
- The average budget of a succesful movie is around 23.4 million with the average revenue around 87.29 million.
- Movies made on higher budgets also produce greater revenues.
